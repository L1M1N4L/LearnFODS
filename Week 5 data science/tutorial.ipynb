```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python

data=load_iris(as_frame=True)
iris = data.data
target=data.target
```

```python
plt.figure(figsize=(10, 6))

scatter = plt.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'], 
                      c=target, s=iris['petal length (cm)']*100, alpha=0.5)

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Sepal Dimensions')

# Add legend
unique_targets = np.unique(target)
species_names = ['setosa', 'versicolor', 'virginica']  # Default Iris species names
if len(unique_targets) != len(species_names):
    species_names = [f'Species {i}' for i in range(len(unique_targets))]
legend1 = plt.legend(handles=scatter.legend_elements()[0],
                     labels=species_names,
                     title="Species",
                     loc="upper left")
plt.gca().add_artist(legend1)

# Add colorbar to show petal length scale
cbar = plt.colorbar()
cbar.set_label('Petal Length (cm)', rotation=270, labelpad=15)
plt.tight_layout()
plt.show()
```

Seaborn

```python

# Set the style (optional)
sns.set_style("whitegrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=iris, 
                x='sepal length (cm)', 
                y='sepal width (cm)', 
                hue=target,  # Color by species
                size='petal length (cm)',  # Size by petal length
                sizes=(20, 200),  # Range of point sizes
                alpha=0.7)

# Customize the plot
plt.title('Iris Sepal Dimensions')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Modify the legend
handles, labels = scatter.get_legend_handles_labels()
species_names = ['setosa', 'versicolor', 'virginica']
legend = plt.legend(handles[:3], species_names, title='Species', loc='upper left')
plt.gca().add_artist(legend)

# Adjust the colorbar
colorbar = scatter.collections[0].colorbar
colorbar.set_label('Petal Length (cm)', rotation=270, labelpad=15)
plt.tight_layout()
plt.show()
```
